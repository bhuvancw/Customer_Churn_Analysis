```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer Churn.csv')
df.head(40)
```

```python
df.describe(include = 'all')
```

### Replacing the blanks with 0 as terure is 0 and no total charges are recorded

```python
df['TotalCharges'] = df['TotalCharges'].str.strip().replace('',"0")
df['TotalCharges'] = df['TotalCharges'].astype("float")
df.info()
```

### Converted 0s and 1s from Senior Citizen column to Yes/ No

```python
def conv(value):
    if value == 1:
        return 'yes'
    else:
        return 'no'
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
df.head(40)
```

```python
ax = sns.countplot(x = df['Churn'])
ax.bar_label(ax.containers[0])
plt.title('Count of Churned Customers')
plt.show()
```

```python
plt.figure(figsize = (4,3))
gb = df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'],autopct = '%.2f%%',startangle = 90, labels = gb.index )
plt.title('% of of Churned Customers')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (4,4))
sns.countplot(x = df['gender'], hue = df['Churn'])
plt.tight_layout()
plt.title('Churn by Gender')
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = 'SeniorCitizen', data = df, hue = 'Churn')
for container in ax.containers:
    ax.bar_label(container)
plt.title('Churn by Senior Citizen')
plt.show()
```

```python
table = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Convert to %
table_pct = table.div(table.sum(1), axis=0) * 100

# Step 3: Plot stacked bar chart
ax = table_pct.plot(kind="bar", stacked=True, figsize=(5,5))

# Step 4: Add % labels inside the bars
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%", label_type="center")

plt.ylabel("Percentage")
plt.title("Churn by Senior Citizen (%)")
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Setup subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15,12))
axes = axes.flatten()  # make it a 1D array for easy iteration

# Loop through each column and plot
for i, col in enumerate(cols):
    ax = sns.countplot(x=df[col], ax=axes[i], hue = df['Churn'])
    axes[i].set_title(f"Count of {col}")
    axes[i].set_xlabel("")  # remove x-axis label (optional)
    axes[i].set_ylabel("Count")

    for container in ax.containers:
        ax.bar_label(container)

# Adjust layout so plots donâ€™t overlap
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (7,4))
ax = sns.countplot(x = df['PaymentMethod'], hue = df['Churn'])
for container in ax.containers:
    ax.bar_label(container)
plt.xticks(rotation = 45)
plt.title('Count of Churned Customers by Payment Method')
plt.show()
```
